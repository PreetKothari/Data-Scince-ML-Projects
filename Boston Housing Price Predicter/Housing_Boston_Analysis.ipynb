## Machine Learning Algorithms for MEDV/Prices Prediction

```python
X = pd.DataFrame(dataset[['RM','LSTAT']])
#rm = pd.DataFrame(dataset['RM'])
#lstat = pd.DataFrame(dataset['LSTAT'])
#X = pd.concat([lstat,rm], axis = 1)
# we can also use 'X = pd.DataFrame(np.c_[dataset['LSTAT'], dataset['RM']], columns = ['LSTAT','RM'])'
Y = dataset['MEDV']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
print("X Train: ", X_train.shape)
print("Y Train: ", Y_train.shape)
print("X Test: ", X_test.shape)
print("Y Test: ", Y_test.shape)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
```

```python
print("Linear Regression Model")
model_lr = pd.DataFrame(X_test)
model_lr['MEDV'] = Y_test
model_lr['Predicted MEDV'] = Y_pred
print(model_lr.head(10))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
```

```python
err = 'MSE: ' + str(round(mse,2)) + ', ' + 'MAE: ' + str(round(mae,2))
print('Linear Regression Model Performance:-- ',err)
```

```python
plt.figure(figsize = (12,9))
ax = sns.regplot(Y_test, Y_pred, marker = 'o', color = 'green')
ax.set_title('Linear Regression', fontsize = 20)
ax.set_xlabel('MEDV or Price', fontsize = 20)
ax.set_ylabel('Predicted Prices', fontsize = 20)
plt.text(35.0, 0.0, err, fontsize=20, bbox=dict(facecolor='green', alpha=0.5))
```

### K - Nearest Neighbour 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors = 1)
knr.fit(X_train, Y_train)
```

```python
Y_pred = knr.predict(X_test) 
```

```python
print("KNN Regression Model")
model_knr = pd.DataFrame(X_test)
model_knr['MEDV'] = Y_test
model_knr['Predicted MEDV'] = Y_pred
print(model_knr.head(10))
```

```python
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
```

```python
err = 'MSE: ' + str(round(mse,2)) + ', ' + 'MAE: ' + str(round(mae,2))
print('KNN Regression Model Performance:-- ',err)
```

```python
plt.figure(figsize = (12,9))
ax = sns.regplot(Y_test, Y_pred, marker = 'o', color = 'blue')
ax.set_title('KNN Regression', fontsize = 20)
ax.set_xlabel('MEDV or Price', fontsize = 20)
ax.set_ylabel('Predicted Prices', fontsize = 20)
plt.text(35.0, 10.0, err, fontsize=20, bbox=dict(facecolor='blue', alpha=0.5))
```

### Gradient Boosting Tree Regression

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(X_train, Y_train)
```

```python
Y_pred = gbr.predict(X_test)
print("Gradient Boost Regression Model")
model_gbr = pd.DataFrame(X_test)
model_gbr['MEDV'] = Y_test
model_gbr['Predicted MEDV'] = Y_pred
print(model_gbr.head(10))
```

```python

mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
```

```python
err = 'MSE: ' + str(round(mse,2)) + ', ' + 'MAE: ' + str(round(mae,2))
print('Gradient Boost Regression Model Performance:-- ',err)
```

```python
plt.figure(figsize = (12,9))
ax = sns.regplot(Y_test, Y_pred, marker = 'o', color = 'red')
ax.set_title('Gradient Boost Regression', fontsize = 20)
ax.set_xlabel('MEDV or Price', fontsize = 20)
ax.set_ylabel('Predicted Prices', fontsize = 20)
plt.text(35.0, 10.0, err, fontsize=20, bbox=dict(facecolor='red', alpha=0.5))
```

### Model Validation 

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
fold = KFold(n_splits=10, shuffle=True, random_state = 2)
print(fold)
```

```python
print("Gradient Boost Regression Cross Validation with KFold:")
kfolds = [2, 3, 4, 5,10]
for kfold in kfolds :
    cv_results = cross_val_score(gbr, X, Y, cv = KFold(kfold, random_state = 0))
    print(kfold,cv_results)
    print("Kfold = ", kfold, ", MAE = ", round(cv_results.mean(), 2), ", Sdev = ",round(cv_results.std(),2))
```

```python
# Model Validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
# Cross validate model with Kfold stratified cross val
# fold = KFold(n_splits=10, shuffle=True, random_state = 2)
# print(fold)

# Modeling step Test differents algorithms
# random_state = 2
regressors = []
regressors.append(LinearRegression())
regressors.append(KNeighborsRegressor())
regressors.append(GradientBoostingRegressor())


regressors = [lr, knr, gbr]
kfolds = [2, 3, 4, 5]
for kfold in kfolds :
    print ("Kfold ", kfold)
    for regressor in regressors :
        cv_results = (cross_val_score(regressor, X, Y, cv = KFold(kfold, random_state = 0)))
        if regressor == lr :
            print("Linear Regression, MAE: ", round(cv_results.mean(), 2))
        elif regressor == knr:
            print("KNN Regression, MAE: ", round(cv_results.mean(), 2))
        else:
            print("Gradient Boost Regression, , MAE: ", round(cv_results.mean(), 2))
```
